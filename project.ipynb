## Practice vs Playing time
**Describe the problem/question:** Do I spend more time practicing my golf swing technique or playing on the course?

**Describe the data:** Practicing time (range, chipping, putting) vs playing time

**Hypothesis t-test type:** Two-sampled, two-tailed, dependent t-test because there are two samples of data, two-tailed because the data is not going in a specific direction, and dependent because each measurement is taken on the same days.

### 5 step hypothesis testing approach:
1. Null and alternative hypotheses: $H_{0}: \mu_{practice} ≠ \mu_{play}$ and $H_{1}: \mu_{practice} > \mu_{play}$
2. Level of significance: $\alpha = 0.05$
3. Test statistic: $$t = \frac{\bar{d}}{s_d / \sqrt{n}}$$
4. Decision rule: Reject $H_{0}$ if $p_{two-tailed} < \alpha = 0.05$
5. $t_{computed} = \frac{\bar{d}}{s_d / \sqrt{n}} = 0.0068$

$p_{two-tailed} = 0.0068 < \alpha = 0.05$ $\Rightarrow \text{Reject } H_{0}$

```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("golf_practice.csv")
df["Total_practice"] = df["Range"] + df["Chipping"] + df["Putting"]
df["Difference"] = df["Total_practice"] - df["Playing"]

mean = df["Difference"].mean()
std = df["Difference"].std()

t_stat, p_val = stats.ttest_rel(df["Total_practice"], df["Playing"])
print(p_val)

alpha = 0.05
if p_val < alpha:
    print("Reject H0")
    if mean > 0:
        print("Significantly more time was spent practicing than playing")
    else:
        print("Significantly more time was spent playing than practicing")
else:
    print("Fail to reject H0, no significant difference between practice and playing time")
```
